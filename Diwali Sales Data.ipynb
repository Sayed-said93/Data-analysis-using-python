```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\DiwaliSalesData.csv",encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.sample(20)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int64')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df=df.rename(columns={'Marital_Status':'marriage'})
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender',data=df)
for y in ax.containers:
    ax.bar_label(y)
```

```python
sales_by_Gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_by_Gen , color='orange')
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for y in ax.containers:
    ax.bar_label(y)
```

```python
age_amount=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=age_amount)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
bx=sns.countplot(x='marriage',data=df)
sns.set(rc={'figure.figsize':(2,5)})
for y in bx.containers:
    bx.bar_label(y)
```

```python
age_s_amount=df.groupby(['marriage','Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='marriage',y='Amount',data=age_s_amount,hue='Gender')
```

```python
bx=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(30,5)})
for y in bx.containers:
    bx.bar_label(y)
```

```python
ocu_amount=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=ocu_amount)
```

```python
bx=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,5)})
for y in bx.containers:
    bx.bar_label(y)
```

```python
pro_amount=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=pro_amount)
```

```python
pro_ord=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=pro_ord,x='Product_ID',y='Orders')
```
